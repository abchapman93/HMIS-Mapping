```python
hmis.columns = [clean_column_name(col) for col in hmis.columns]
year_cols = dict()
year = None
for col in hmis.columns[1:]:
    if col.endswith("_response_category"):
        year = col[:4]
        if re.search("20\d\d_v\d", col):
            year = col[:7]
    
    year_cols.setdefault(year, list([hmis.columns[0]]))
    year_cols[year].append(col)
    
```

```python
year_cols
```

```python
df_years = list()
final_cols = ['data_element_field', 'hmis_version', 'response_category', 'dependent_field',
       'dependent_response', 'type', 'location', 'dsd_source', 'xml_value',
       'collection_point', 'change']

for i, (year, cols) in enumerate(year_cols.items()):
    df_year = hmis[year_cols[year]].copy()
    # Remove year and suffixes from column names
    new_cols = [df_year.columns[0]]
    for col in df_year.columns[1:]:
        new_col = col
        if col.startswith(f"{year}_"):
            new_col = new_col[len(year)+1:]
        elif (match := re.search("\.\d+$", col)) :
            new_col = new_col[:match.span()[0]]
        elif col.endswith("_change"):
            new_col = "change"
        new_cols.append(new_col)
    df_year.columns = new_cols
    df_year["hmis_version"] = year

    if (year == "2014_v2"):
        print(i)
    
    
    df_years.append(df_year)
```

```python
def get_mapping(descr):
    x = descr.lower()
    if x in ("client doesn’t know", "client prefers not to answer", "client refused", "data not collected"):
        return "Missing"
    if x.startswith("emergency shelter"):
        return "Sheltered"
    if re.search("place not meant for (human )?habitation", x):
        return "Unsheltered"
    if x in ("safe haven", "interim housing"):
        return "Sheltered"
    if "foster" in x or "group home" in x:
        return "Institutitional"
    if re.search("hospital|nursing home|treatment facility", x):
        return "Healthcare"
    if re.search("jail|prison", x):
        return "Institutional"
    if re.search("(owned|rental) by client", x):
        return "Housed"
    if re.search("[hm]otel", x):
        return "Sheltered"
    if re.search("halfway house|residental project|safe haven", x):
        return "Sheltered"
    if re.search("host home", x):
        return "Housed"
    if "permanent housing" in x:
        return "Housed"
    if re.search("staying or living in a (friend|family)", x):
        return "Housed"
    if re.search("transitional housing", x):
        return "Sheltered"
    if re.search("do(es)?n't know|refused|other", x):
        return "Missing"
    
    return ""
```

```python
ls["category"] = ls["description"].fillna("").apply(get_mapping)
ls[["description", "category"]].drop_duplicates()
```

```python
def get_mapping_dest(descr):
    x = descr.lower()
    if x in ("client doesn’t know", "client prefers not to answer", "client refused", "data not collected"):
        return "Missing"
    if x.startswith("emergency shelter"):
        return "Sheltered"
    if re.search("place not meant for (human )?habitation", x):
        return "Unsheltered"
    if x in ("safe haven", "interim housing"):
        return "Sheltered"
    if "foster" in x or "group home" in x:
        return "Institutitional"
    if re.search("hospital|nursing home|treatment facility", x):
        return "Healthcare"
    if re.search("jail|prison", x):
        return "Institutional"
    if re.search("(owned|rental) by client", x):
        return "Housed"
    if re.search("[hm]otel", x):
        return "Sheltered"
    if re.search("halfway house|residental project", x):
        return "Sheltered"
    if re.search("host home", x):
        return "Housed"
    if "permanent housing" in x:
        return "Housed"
    if re.search("staying or living (in a|with) (friend|family)", x):
        return "Housed"
    if re.search("transitional housing", x):
        return "Sheltered"
    if re.search("do(es)?n't know|refused|other", x):
        return "Missing"
    if re.search("deceased", x):
        return "Deceased"
    if x.endswith("HOPWA PH".lower()):
        return "Housed"
    if x.endswith("HOPWA TH".lower()):
        return "Sheltered"
    if x == "no exit interview completed":
        return "Missing"
    return ""
```

# Education

```python
import re
```

```python
educ = long[long["data_element_field"].fillna("").str.contains("(grade complete)", flags=re.IGNORECASE, regex=True)].copy()
educ = educ[~educ["value"].isna()]
```

```python
educ["data_element_field"] = "Education"
```

```python
def get_mapping_education(descr):
    x = descr.lower()
    if descr == "Less than Grade 5":
        return "Less than high school"
    if re.search("grades \d\-\d", x):
        return "Less than high school"
    if x == "grade 12":
        return "High school diploma"
    if descr == "GED":
        return "High school diploma"
    if descr == "School program does not have grade levels":
        return "Missing"
    if descr == "Some college":
        return descr
    if re.search("(client doesn’t know|refused|not collected|not to answer)", x):
        return "Missing"
    if re.search("(associate's|bachelor's|graduate) degree", x):
        return "College degree"
    if descr == "Vocational Certification":
        return "College degree"
    return ""
```

```python
educ["category"] = educ["description"].fillna("").apply(get_mapping_education)
educ[["description", "value", "category"]].drop_duplicates()
```

## Employment Status

```python
employ_status = long[long["data_element_field"].fillna("").str.contains("employment status - employed", flags=re.IGNORECASE)].copy()
employ_status = employ_status[~employ_status["value"].isna()]
employ_status["data_element_field"] = "Employment Status"
```

```python
def get_mapping_employment_status(descr):
    if descr == "Yes":
        return "Employed"
    if descr == "No":
        return "Unemployed"
    if re.search("doesn't know|refused|not collected|prefers not to answer", descr, re.IGNORECASE):
        return "Missing"
    return
```

```python
employ_status
```

```python
employ_status["category"] = employ_status["description"].fillna("").apply(get_mapping_employment_status)
employ_status[["description", "value", "category"]].drop_duplicates()
```

## Employment Type

```python
employ_type = long[long["data_element_field"].fillna("").str.contains("employment status - type of employment", flags=re.IGNORECASE)].copy()
employ_type = employ_type[~employ_type["value"].isna()]
employ_type["data_element_field"] = "Employment Type"

# This is a 'dependent response', so the descriptions are in a different column
employ_type["description"] = employ_type["dependent_response"]
```

```python
def get_mapping_employment_type(descr):
    if descr in ("Full-Time", "Part-time"):
        return descr
    if descr.startswith("Seasonal"):
        return "Part-time"
    
    return
```

```python
employ_type["category"] = employ_type["description"].fillna("").apply(get_mapping_employment_type)
employ_type[["description", "value", "category"]].drop_duplicates()
```

```python
employ_type = select_final_columns(employ_type)
```
